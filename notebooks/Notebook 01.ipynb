Updating list with new version

```python
# Filter diabetes from its own topic
df_diabetes = df[df['Topic'] == 'Diabetes'].copy()

# Filter obesity and inactivity from Nutrition topic by specific question
df_obesity = df[df['Question'] == 'Obesity among adults aged >= 18 years'].copy()

df_inactivity = df[df['Question'] == 'No leisure-time physical activity among adults aged >= 18 years'].copy()

# Add a label column to each so we know which is which
df_diabetes['indicator'] = 'diabetes'
df_obesity['indicator'] = 'obesity'
df_inactivity['indicator'] = 'inactivity'

# Combine all three into one dataframe
df_filtered = pd.concat([df_diabetes, df_obesity, df_inactivity], ignore_index=True)

print("Filtered shape:", df_filtered.shape)
print(df_filtered['indicator'].value_counts())
```

```python
# Rerunning cells since the data I am looking for is now visible after some investigation work 
df_pivot = df_filtered.pivot_table(
    index=['LocationAbbr', 'YearStart'],
    columns='indicator',
    values='DataValue',
    aggfunc='mean'
).reset_index()

df_pivot.columns.name = None
df_pivot = df_pivot.rename(columns={
    'diabetes': 'diabetes_prevalence',
    'obesity': 'obesity_prevalence',
    'inactivity': 'inactivity_prevalence'
})

df_pivot = df_pivot.dropna()
print(df_pivot.shape)
print(df_pivot.head())
```

```python
# Convert DataValue to numeric (any non-numeric values become NaN)
df_filtered['DataValue'] = pd.to_numeric(df_filtered['DataValue'], errors='coerce')

print("DataValue dtype:", df_filtered['DataValue'].dtype)
print("Sample values:", df_filtered['DataValue'].dropna().head())
```

```python
# Rerunning cells after converting to float type 
df_pivot = df_filtered.pivot_table(
    index=['LocationAbbr', 'YearStart'],
    columns='indicator',
    values='DataValue',
    aggfunc='mean'
).reset_index()

df_pivot.columns.name = None
df_pivot = df_pivot.rename(columns={
    'diabetes': 'diabetes_prevalence',
    'obesity': 'obesity_prevalence',
    'inactivity': 'inactivity_prevalence'
})

df_pivot = df_pivot.dropna()
print(df_pivot.shape)
print(df_pivot.head())
```

```python
sns.set(style="whitegrid")

pair_plot = sns.pairplot(
    df_pivot[['diabetes_prevalence', 'obesity_prevalence', 'inactivity_prevalence']],
    diag_kind='kde',
    plot_kws={'alpha': 0.5, 'color': 'steelblue'}
)

pair_plot.fig.suptitle("Pair Plot: Diabetes, Obesity, Inactivity", y=1.02)
plt.savefig(viz_path + "eda_pairplot.png", bbox_inches='tight', dpi=150)
plt.show()
print("Pair plot saved!")
```

```python
corr_matrix = df_pivot[['diabetes_prevalence', 'obesity_prevalence', 'inactivity_prevalence']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="RdBu_r",
    center=0,
    square=True,
    linewidths=0.5
)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig(viz_path + "eda_correlation_heatmap.png", dpi=150)
plt.show()
print("Heatmap saved!")
```
